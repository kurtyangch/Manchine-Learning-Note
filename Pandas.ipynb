```python
# column from series 新增欄位
seats_column = pd.Series([5, 5, 5 ,5 ,5])
car_sales["Seats"] = seats_column
```

```python
car_sales["Seats"].fillna(5, inplace = True)
```

```python
# column from Python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7,7.6,8.7,3.1,4.5]
car_sales["Fuel_per_100KM"] = fuel_economy
```

```python
car_sales["Total fuel used"] = car_sales["Odometer (KM)"] / 100 * car_sales["Fuel_per_100KM"]
```

```python
# create a column from a single value
car_sales["Number of wheels"] = 4
```

```python
# apply() --> assign a function to column, lambda x : x/1.6(對 x 做以下的指令)
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x:x/1.6)
```
